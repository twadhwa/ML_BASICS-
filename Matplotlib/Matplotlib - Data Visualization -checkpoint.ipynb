## Working of the matplotlib 

```python
# 1) Prepare data 
x=[1,2,3,4]
y=[11,22,33,44]

# 2) Create plot 
fig ,ax = plt.subplots(figsize=(5,5))

# figsize are the dimensions of the figure 

# 3) Display plot 
ax.plot(x,y)

# 4) Customising plot 
ax.set(title = "Simple Plot", 
      xlabel = "X axis",
      ylabel = "Y axis")

# 5) saving the plot for the future use 
fig.savefig("sample-Plot.png")
       
       
```

### Which method use for plotting ?? (pyplot vs the matplotlib object oriented method)

* For plotting fast we can use pyplot method 
* For plotting something advance use the oo method 

```python
heart 
```

```python
over_50 = heart[heart["age"] > 50]
over_50
```

```python
(len(over_50)/ len(heart)) * 100
```

```python
# So as we can see it is almost the 70% of the original data set 
```

```python
# Pyplot method 
over_50.plot(kind = "scatter"
             , x = "age", 
             y = "chol", 
             c = "target");
```

```python
fig ,ax = plt.subplots (figsize = (10,5 ))
over_50.plot(kind = "scatter",
            x ="age", 
            y = "chol", 
             c = "target",
            ax =ax ); 
```

```python
 fig ,ax = plt.subplots (figsize = (10,5 ))
over_50.plot(kind = "scatter",
            x ="age", 
            y = "chol", 
             c = "target",
            ax =ax ); 
ax.set_xlim([45,100]);
# This functin will change the limit from 45 to 100 
```

```python
# The above method is oo method mixed with the pyplot method 
```

```python
# Now we will make plot using the oo method from the scratch
```

```python
# 1 ) Creating the plot 
fig ,ax = plt.subplots(figsize = (10,5 )) 
 
# 2) plotting the data on the curve 
scatter = ax.scatter (x= over_50 [ "age"], 
                     y = over_50 [ "chol"],
                     c = over_50 ["target"]);

# Customize the plot 
ax.set(title = "Heart Disease", 
      xlabel = "Age",
      ylabel = " chlorestrol");

# Creating a legend 
ax.legend(*scatter.legend_elements(), title = "target");

# Adding a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle = "--")
```

```python
# As you can see the object oriented method is faar better than pyplot method 
```

```python
# Creating a subplot of chol ,age ,thalach 
```

```python
over_50
```

```python
fig , ((ax_1),(ax_2) )=plt.subplots (nrows = 2 , 
                                     ncols = 1 , figsize = (10,10
                                                        ))

scatter_1 = ax_1.scatter(x = over_50["age"], 
                        y = over_50["chol"], 
                        c = over_50["target"]);
scatter_2 =ax_2.scatter(x= over_50["age"],
                        y = over_50["thalach"], 
                        c= over_50["target"]);

ax_1.set (title =  "age vs chol", 
              xlabel = "age ", 
              ylabel = "cholestrol", 
              );
ax_2.set(title = "age vs thrach", 
             xlabel = "age", 
             ylabel = "thrach ");

ax_1.legend(*scatter.legend_elements(), title = "target");
ax_1.legend(*scatter.legend_elements(), title = "target");


ax_1.axhline(over_50["chol"].mean(), linestyle = "--");

ax_2.axhline(over_50["thalach"].mean(), linestyle = "--");

# Adding title to the whole fig 
fig.suptitle ("Heart Disease analysis"); 
```
